```python
import pandas as pd

diabetes_df = pd.read_csv('diabetes_model.csv', encoding='ascii')

print(diabetes_df.head())
print('Loaded diabetes_model.csv successfully.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = diabetes_df

# Set the style for the plots
sns.set(style="whitegrid")

# List of variables to analyze
variables = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(variables), ncols=2, figsize=(12, 4 * len(variables)))

# Loop through each variable to create histograms and box plots
for i, var in enumerate(variables):
    # Histogram
    sns.histplot(data[var], bins=30, kde=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram of {var}')
    axes[i, 0].set_xlabel(var)
    axes[i, 0].set_ylabel('Frequency')

    # Box Plot
    sns.boxplot(x=data[var], ax=axes[i, 1])
    axes[i, 1].set_title(f'Box Plot of {var}')
    axes[i, 1].set_xlabel(var)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python

# Set the style for the plots
sns.set(style="whitegrid")

# Create a scatter plot for selected pairs of variables
plt.figure(figsize=(12, 8))

# Scatter plot: Glucose vs. BMI
plt.subplot(2, 2, 1)
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=data, palette='Set1')
plt.title('Scatter Plot: Glucose vs. BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')

# Scatter plot: Insulin vs. Skin Thickness
plt.subplot(2, 2, 2)
sns.scatterplot(x='Insulin', y='SkinThickness', hue='Outcome', data=data, palette='Set1')
plt.title('Scatter Plot: Insulin vs. Skin Thickness')
plt.xlabel('Insulin')
plt.ylabel('Skin Thickness')

# Scatter plot: Age vs. Glucose
plt.subplot(2, 2, 3)
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=data, palette='Set1')
plt.title('Scatter Plot: Age vs. Glucose')
plt.xlabel('Age')
plt.ylabel('Glucose')

# Scatter plot: BMI vs. Diabetes Pedigree Function
plt.subplot(2, 2, 4)
sns.scatterplot(x='BMI', y='DiabetesPedigreeFunction', hue='Outcome', data=data, palette='Set1')
plt.title('Scatter Plot: BMI vs. Diabetes Pedigree Function')
plt.xlabel('BMI')
plt.ylabel('Diabetes Pedigree Function')

# Adjust layout
plt.tight_layout()
plt.show()

# Create a correlation matrix
correlation_matrix = data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix')
plt.show()
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Pair plot for selected variables
pairplot_vars = ['Pregnancies', 'Glucose', 'BMI', 'Age', 'Outcome']
sns.pairplot(data[pairplot_vars], hue='Outcome', palette='Set1')
plt.suptitle('Pair Plot of Selected Variables', y=1.02)
plt.show()

# 3D Plot: Glucose, BMI, and Age
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot in 3D
scatter = ax.scatter(data['Glucose'], data['BMI'], data['Age'], c=data['Outcome'], cmap='Set1', marker='o')

# Labels and title
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('3D Scatter Plot: Glucose, BMI, and Age')

# Add a legend
legend1 = ax.legend(*scatter.legend_elements(), title="Outcome")
ax.add_artist(legend1)

plt.show()
```
